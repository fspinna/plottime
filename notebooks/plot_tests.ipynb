```python
from plottime.plot import plot, plot_saliency
import awkward as ak
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate multivariate time series
```

```python
Y = [[np.sin(np.arange(0, 40, 0.1)), np.cos(np.arange(0, 70, 0.2))]]
Y = ak.Array(Y)
```

```python
plot(Y)
plt.show()
```

```python
# Generate a custom x-axis
```

```python
X = ak.Array([[np.arange(0, 800, 2), np.arange(0, 350, 1)]])
```

```python
plot(Y, X, sharex=False)
plt.show()
```

```python
# Generate a saliency map
```

```python
S = ak.Array([[np.arange(0, 400, 1), np.arange(-175, 175, 1)**2]])
```

```python
plot_saliency(Y=Y, X=X, S=S, alpha=0.9, cmap="coolwarm", sharex=False)
plt.show()
```
